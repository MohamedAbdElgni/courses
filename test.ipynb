```python
import json
import requests
import base64

with open("conf.json", "r") as config_file:
    config = json.load(config_file)

client_id = config["client_id"]
client_secret = config["client_secret"]
# Replace with your actual client ID and client secret

# Base URL for Udemy Affiliate API
base_url = "https://www.udemy.com/api-2.0"

# Endpoint for courses

# Create the authentication header with base64-encoded client ID and client secret
auth_string = f"{client_id}:{client_secret}"
auth_header = base64.b64encode(auth_string.encode()).decode("utf-8")

# Add the Authorization header to the request
headers = {"Authorization": f"Basic {auth_header}"}
#for price list
# courses_endpoint = "/pricing/?course_ids=5498946"
courses_endpoint = "/courses/?page=1&page_size=100&search=Most-Popular&new&trending&ordering=relevance"
# Make a GET request to retrieve course data
response = requests.get(base_url + courses_endpoint, headers=headers)

# Check if the request was successful (HTTP status code 200)

course_data = response.json()
    # You now have the course data in the course_data variable
    # now we ptint it in organized way

print(json.dumps(course_data, indent=4, sort_keys=True))
```

```python
from bs4 import BeautifulSoup
url = "https://www.udemy.com/sitemap/"
req = requests.get(url)
html = req.text
soup = BeautifulSoup(html, 'html.parser')

# Find all elements with class="row row-gap-md"
categories = soup.find_all('b')

for category in categories:
    main_category = category.text.strip()
    subcategories = category.find_next('ul').find_all('a')

    print(f'Main Category: {main_category}')

    for subcategory in subcategories:
        print(f'Subcategory: {subcategory.text}')

    print()
# Iterate through each row
# for row in rows:
#     # Check if the current element is a main category
#     if row.find('b'):
#         main_category = row.find('b').text
#         print(f"\nMain Category: {main_category}\n")

#         # Extract and print the text from the list within the row (subcategories)
#         subcategories = row.find('ul').find_all('a')
#         for subcategory in subcategories:
#             print(f"Subcategory: {subcategory.text}")
```
